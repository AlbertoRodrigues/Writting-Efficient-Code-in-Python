# Usar sets no dia a dia pode trazer algumas vantagens em relação a listas. Esse tipo de objeto é mais rápido em relação a listas

```python
list_a = ['Bulbasaur', 'Charmander', 'Squirtle']
list_b = ['Caterpie', 'Pidgey', 'Squirtle']
```

```python
set_a = set(list_a)
print(set_a)
```

```python
set_b = set(list_b)
print(set_b)
```

```python
set_a.intersection(set_b)
```

```python
set_a.difference(set_b)
```

```python
set_a.symmetric_difference(set_b)
```

```python
set_a.union(set_b)
```

# Aqui podemos ver que usando o objeto set, o uso de verificar se algum elemento pertence a uma lista/tupla/set, a velocidade de execução usando set é a mais rápidas de todas.
